## 백터화를 위한 DF 생성

```python
df['통합정보'] = df.apply(lambda row: f"""
관광지명: {row['관광지명']}
도로명주소: {row['소재지도로명주소'] or '정보 없음'}
지번주소: {row['소재지지번주소']}
관광지소개: {row['관광지소개']}
공공편익시설정보: {row['공공편익시설정보'] or '정보 없음'}
숙박시설정보: {row['숙박시설정보'] or '정보 없음'}
운동및오락시설정보: {row['운동및오락시설정보'] or '정보 없음'}
접객시설정보: {row['접객시설정보'] or '정보 없음'}
지원시설정보: {row['지원시설정보'] or '정보 없음'}
수용인원: {row['수용인원수']}
주차가능수: {row['주차가능수']}
관리기관 및 연락처: {row['관리기관명']} - {row['관리기관전화번호']}
위도, 경도: [{row['위도']}, {row['경도']}]
면적: {row['면적']}
""", axis=1)
```

```python
# NaN 및 빈값을 '없음'으로 처리하는 함수
def format_value(value):
    return '없음' if pd.isna(value) or value == '' else value

# 태그 추출 함수 (최적화)
def extract_tag(row):
    # 도로명주소와 지번주소 추출
    road_address = row['소재지도로명주소']
    lot_address = row['소재지지번주소']
    
    # 도로명 주소가 유효하고 2개 이상의 단어로 구성된 경우 사용
    if pd.notna(road_address) and len(road_address.split()) >= 2:
        return ' '.join(road_address.split()[:2])
    
    # 지번주소로 대체 (유효성 체크)
    if pd.notna(lot_address) and lot_address.strip() != '':
        return ' '.join(lot_address.split()[:2])
    
    # 둘 다 유효하지 않을 경우
    return '없음'
```

```python
newDF = pd.DataFrame()
```

```python
newDF['이름'] = df.apply(lambda row: f"""{format_value(row['관광지명'])}""", axis=1)
newDF['내용'] = df.apply(lambda row: f"""
관광지명: {format_value(row['관광지명'])}
도로명주소: {format_value(row['소재지도로명주소'])}
지번주소: {format_value(row['소재지지번주소'])}
관광지소개: {format_value(row['관광지소개'])}
공공편익시설정보: {format_value(row['공공편익시설정보'])}
숙박시설정보: {format_value(row['숙박시설정보'])}
운동및오락시설정보: {format_value(row['운동및오락시설정보'])}
접객시설정보: {format_value(row['접객시설정보'])}
지원시설정보: {format_value(row['지원시설정보'])}
수용인원: {format_value(row['수용인원수'])}
주차가능수: {format_value(row['주차가능수'])}
관리기관 및 연락처: {format_value(row['관리기관명'])} - {format_value(row['관리기관전화번호'])}
위도, 경도: [{format_value(row['위도'])}, {format_value(row['경도'])}]
면적: {format_value(row['면적'])}
""", axis=1)

# ✅ 태그 컬럼 생성
newDF['태그'] = df.apply(lambda row: extract_tag(row), axis=1)
```
